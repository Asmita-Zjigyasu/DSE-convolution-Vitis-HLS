```python
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
df=pd.read_csv("consolidated_report (copy).csv")
```

```python
df.describe()
```

```python
df.info()
```

```python
class_labels=['Estimated Clock Period','BRAM_18K percentage','DSP percentage','FF percentage','LUT percentage']
flag=True
base="AnalysisCSV"
min_dict={}
for i in class_labels:
    classes=list(set(df[i]))
    classes.sort()
    min_dict[i]=classes[0]
    dir=os.path.join(base,i.replace(" ","_"))
    os.system("mkdir -p "+dir)
    for j in classes:
        df2=df[df[i]==j]
        df2.to_csv(os.path.join(dir,str(j))+".csv")
        
```

```python
class_labels=['Estimated Clock Period','BRAM_18K percentage']
print(min_dict)
df2=df
for i in class_labels:
    df2=df2[df2[i]==min_dict[i]]
df2.describe()
```
